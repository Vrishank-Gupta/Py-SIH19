```python
import urllib.request, json
import requests
def relevantWord(keyword):
    a = keyword + ""
    keyword = a.replace(' ','%20')
    page = "https://relatedwords.org/api/related?term=" + keyword


    r = requests.get(page)
    
    result = r.json()
    word_list = []
    score_list = []
    for i in range(50):
        word_list.append(result[i]["word"])
        score_list.append(result[i]["score"])
        print(result[i])

    return word_list,score_list
```

```python
word_list,score_list = relevantWord("kidney")
```

```python
def wordList(list_of_words):
    size = len(list_of_words)
    words = []
    scores = []
    for i in range(size):
        word_list,score_list = relevantWord(list_of_words[i])
        words.append(word_list)
        print("\n\n\n\n\n\n\n\n\n")
        scores.append(score_list)
    
    return words,scores
```

```python
words,score = wordList(["arthiritis","catheter ablation", "Arrhythmia","renal","histocompatibility","penicillin","neoangiogenesis","arteries"])
# print(words,score)
```
